## Assignment 1: Basic If-Else Statement

**Task:** Write a Python program that asks the user to input their age and prints whether they are eligible to vote (18 years or older) or not.

```python
age = int(input("Enter your age: "))

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

## Assignment 2: Nested If-Else

**Task:** Write a Python program that checks if a user’s input number is positive, negative, or zero.
- If the number is positive, print 'Positive Number'
- If the number is negative, print 'Negative Number'
- If the number is zero, print 'Zero'

```python
num = float(input("Enter a number: "))

if num > 0:
    print("Positive Number")
elif num < 0:
    print("Negative Number")
else:
    print("Zero")
```

## Assignment 3: Conditional Expressions

**Task:** write a program to determine if a number entered by the user is even or odd. The program should print 'Even Number' or 'Odd Number' based on the input.

```python
n = int(input("Enter a number: "))

if n % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

## Assignment 4: Age Group Classification

**Task:** Write a Python program that takes the user’s age and prints which age group they belong to:
- 'Child' for age 0–12
- 'Teenager' for age 13–19
- 'Adult' for age 20–64
- 'Senior' for age 65 and above

```python
age = int(input("Enter your age: "))

if 0 <= age <= 12:
    print("Child")
elif 13 <= age <= 19:
    print("Teenager")
elif 20 <= age <= 64:
    print("Adult")
elif age >= 65:
    print("Senior")
else:
    print("Invalid age")
```

# Python Programming Assignments - Loops and Functions

## Assignment 1: Sum of Natural Numbers

**Task:** Write a Python function `sum_of_natural_numbers(n)` that returns the sum of all natural numbers up to `n`. Use a loop inside the function.

```python
def sum_of_natural_numbers(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total


num = int(input("Enter a number: "))
print("Sum =", sum_of_natural_numbers(num))
```

## Assignment 2: Factorial of a Number

**Task:** Write a Python function `factorial(n)` that calculates and returns the factorial of `n`. Use a loop to calculate the factorial.

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact

num = int(input("Enter a number: "))
print("Factorial =", factorial(num))
```

## Assignment 3: Multiplication Table

**Task:** Write a Python function `multiplication_table(n)` that prints the multiplication table of `n` up to 10.

```python
def multiplication_table(n):
    for i in range(1, 11):
        print(n, "x", i, "=", n * i)

num = int(input("Enter a number: "))
multiplication_table(num)
```
