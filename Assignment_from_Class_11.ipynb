```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score, confusion_matrix
```

```python
# Change path if needed
df = pd.read_csv("/content/Housing.csv")

print(df.head())
```

```python
df = df.dropna()

binary_cols = ['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
for col in binary_cols:
    df[col] = df[col].map({'yes':1,'no':0})

df['furnishingstatus'] = df['furnishingstatus'].map({
    'furnished':2,
    'semi-furnished':1,
    'unfurnished':0
})

print("Preprocessing done")
print(df.head())
```

```python
X = df.drop("price", axis=1)

y_reg = df["price"]

median_price = df["price"].median()
y_clf = (df["price"] > median_price).astype(int)

print("Median price:", median_price)
print("Class counts:\n", y_clf.value_counts())
```

task 1


```python
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X, y_reg, test_size=0.30, random_state=42)

X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X, y_clf, test_size=0.30, random_state=42)

print("Train size:", X_train_reg.shape[0])
print("Test size:", X_test_reg.shape[0])
```

```python
lr = LinearRegression()
lr.fit(X_train_reg, y_train_reg)

y_pred_lr = lr.predict(X_test_reg)

mae = mean_absolute_error(y_test_reg, y_pred_lr)
rmse = np.sqrt(mean_squared_error(y_test_reg, y_pred_lr))

print("Linear Regression (70–30)")
print("MAE:", mae)
print("RMSE:", rmse)
```

```python
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train_clf, y_train_clf)

y_pred_log = log_model.predict(X_test_clf)

acc_log = accuracy_score(y_test_clf, y_pred_log)
cm_log = confusion_matrix(y_test_clf, y_pred_log)

print("Logistic Regression")
print("Accuracy:", acc_log)
print("Confusion Matrix:\n", cm_log)
```

task 2

```python
svm_linear = SVC(kernel="linear")
svm_linear.fit(X_train_clf, y_train_clf)

y_pred_linear = svm_linear.predict(X_test_clf)
acc_linear = accuracy_score(y_test_clf, y_pred_linear)

print("SVM Linear Accuracy:", acc_linear)
```

```python
svm_rbf = SVC(kernel="rbf")
svm_rbf.fit(X_train_clf, y_train_clf)

y_pred_rbf = svm_rbf.predict(X_test_clf)
acc_rbf = accuracy_score(y_test_clf, y_pred_rbf)

print("SVM RBF Accuracy:", acc_rbf)
```

```python
print("========== FINAL RESULT ==========")
print("Linear Regression RMSE:", rmse)
print("Logistic Regression Accuracy:", acc_log)
print("SVM Linear Accuracy:", acc_linear)
print("SVM RBF Accuracy:", acc_rbf)

if acc_rbf > acc_linear:
    print("RBF kernel performs better")
else:
    print("Linear kernel performs better")
```
